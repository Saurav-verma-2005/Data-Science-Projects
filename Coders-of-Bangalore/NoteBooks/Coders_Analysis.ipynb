```python
# # Coders of Bangalore: Unstructured Data Parser
# ### Analyzing the Instagram Tech Ecosystem
# 
# **Goal:** Extract structured insights from raw, multi-line text dumps of 
# social media profiles using pure Python logic.

import json

# %% [markdown]
# ## 1. Load Raw Data
# Reading the unstructured text dump.

def load_data(filename):
    with open(filename, "r", encoding='utf-8') as f:
        return f.read()

data = load_data("finaldata.txt")

# %% [markdown]
# ## 2. Parsing & Normalization Logic
# This function handles complex string splitting and converts social media 
# shorthand (K for thousands, M for millions) into integers.

def parse_chunk(chunk): 
    chunk = chunk.strip()
    sep_chunk = chunk.split('\n')
    username = sep_chunk[0]
    
    # Clean numerical strings (e.g., "1.2K followers" -> 1200)
    def to_int(text):
        num_part = text.split(" ")[0].replace(",", "").replace("K", "").replace("M", "")
        val = float(num_part)
        if "K" in text: return int(val * 1000)
        if "M" in text: return int(val * 1000000)
        return int(val)

    no_of_posts = int(sep_chunk[1].split(" post")[0].replace(",", ""))
    no_of_followers = to_int(sep_chunk[2])
    no_of_following = to_int(sep_chunk[3])
    
    name = sep_chunk[4]
    type_of_page = sep_chunk[5] if len(sep_chunk) > 5 else "Unknown"
    bio = "\n".join(sep_chunk[6:]) if len(sep_chunk) > 5 else ""
    
    return {
        "username": username, 
        "no_of_posts": no_of_posts, 
        "no_of_followers": no_of_followers, 
        "no_of_following": no_of_following, 
        "name": name, 
        "type_of_page": type_of_page, 
        "bio": bio
    }

# Split text into chunks and parse
chunks = [c for c in data.split("\n\n") if len(c) > 3]
all_profiles = [parse_chunk(c) for c in chunks]

# %% [markdown]
# ## 3. Insights Extraction (EDA)

# Q1: Profile with maximum posts
max_post_user = max(all_profiles, key=lambda x: x['no_of_posts'])
print(f"Top Poster: {max_post_user['username']} with {max_post_user['no_of_posts']} posts")

# Q2: Profile with maximum followers
max_foll_user = max(all_profiles, key=lambda x: x['no_of_followers'])
print(f"Top Influencer: {max_foll_user['username']} with {max_foll_user['no_of_followers']} followers")

# Q3: Unique Categories
categories = {p['type_of_page'] for p in all_profiles}
print(f"Total Unique Community Categories: {len(categories)}")
```
